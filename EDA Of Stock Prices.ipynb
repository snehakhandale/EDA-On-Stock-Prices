```python
df['Date'] = pd.to_datetime(df['Date'])
```

```python
df['Date']
```

```python
# Calculate the difference between 'Close' and 'Open' to determine the days where 'Close' > 'Open'
```

```python
df['Close_Open_Diff'] = df['Close'] - df['Open']
df['Close_Open_Diff']
```

```python
# Calculate the difference between 'High' and 'Low' for date-wise analysis
df['High_Low_Diff'] = df['High'] - df['Low']
df['High_Low_Diff']
```

```python
# 1. Date wise analysis with high and low comparison

plt.figure(figsize=(10, 6))
plt.plot(df['Date'], df['High'], label='High')
plt.plot(df['Date'], df['Low'], label='Low')
plt.xlabel('Date')
plt.ylabel('Price')
plt.title('Date wise analysis - High vs Low')
plt.legend()
plt.show()
```

```python
#2. Days where Close > Open:
plt.figure(figsize=(8, 6))
df['Close_Open_Diff'].apply(lambda x: 1 if x > 0 else 0).value_counts().plot(kind='bar')
plt.title('Days where Close > Open')
plt.xlabel('Close > Open')
plt.ylabel('Count')
plt.xticks([0, 1], ['False', 'True'], rotation=0)
plt.show()
```

```python
#3. Volume related to high/low:
plt.figure(figsize=(8, 6))
plt.scatter(df['Volume'], df['High'], label='High')
plt.scatter(df['Volume'], df['Low'], label='Low')
plt.xlabel('Volume')
plt.ylabel('Price')
plt.title('Volume related to High/Low')
plt.legend()
plt.show()
```

```python
#3. Stock opening duration:
plt.figure(figsize=(8, 6))
df['Close_Open_Diff'].plot(kind='hist', bins=20)
plt.title('Stock Opening Duration Distribution')
plt.xlabel('Duration (Close - Open)')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.plot(df['Date'], df['Close'], label='Close')
plt.plot(df['Date'], df['Adj Close'], label='Adj Close')
plt.xlabel('Date')
plt.ylabel('Price')
plt.title('Close vs Adj-Close Prices Over Time')
plt.legend(bbox_to_anchor=(1.2, 1), loc='upper right')
plt.show()
```
